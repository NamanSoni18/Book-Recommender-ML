```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
```

```python
ratings.head()
```

```python
ratings.rename(columns = {
    "User-ID" : "user_id",
    "Book-Rating" : "rating"
}, inplace=True)
```

```python
ratings.head()
```

```python
ratings.head()
```

```python
ratings.rename(columns = {
    "id" : "user_id",
    "Book-Rating" : "rating"
}, inplace=True)
```

```python
ratings.head()
```

```python
x = ratings['user_id'].value_counts() > 250
```

```python
y = x[x].index
```

```python
y
```

```python
y.shape
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_books.head()
```

```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
final_rating = ratings_with_books.merge('num-rating', on='title')
```

```python
num_rating.rename(columns = {"rating" : "num_of_rating"}, inplace=True)
```

```python
final_rating = ratings_with_books.merge(num_rating, on='title')
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 60]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[246,:].values.reshape(1,-1), n_neighbors = 6)
```
